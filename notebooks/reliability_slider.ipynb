# TAST Reliability Slider

Dial `victors_reliability` from 0.0 (maximal skepticism — only qualitative survives) to 1.0 (full quantitative Bayesian update).

Run the cell below after changing the slider value.

```python
import sys
sys.path.insert(0, "../model")
from bayesian_core import load_streams, apply_reliability, bayes_update, RAW_PRIORS, HYPOTHESES, print_surviving

# --- SLIDER ---
reliability = 0.0   # change this value between 0.0 and 1.0
# --------------

streams = load_streams()
if reliability < 0.05:
    print("*** MAXIMAL SKEPTICISM ***")
    print_surviving()
else:
    scaled = apply_reliability(streams, reliability)
    post = bayes_update(scaled, RAW_PRIORS)
    print(f"victors_reliability = {reliability:.2f}")
    for h in HYPOTHESES:
        print(f"  {h}: {post[h]:6.1%}")
    print('\nConditioning: "If we treat the recorded totals as approximately accurate...')
    print('if we do not, the quantitative claims are undefined."')
```
